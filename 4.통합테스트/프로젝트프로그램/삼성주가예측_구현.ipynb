```python
model_samsung = sam_.copy()
```

```python
# 보조지표 추가
# 1) 이평선(SMA, EMA, WMA) (w = 5,10,15,20,30,60,120)
model_samsung['ma_5'] = ta.SMA(model_samsung.Close, timeperiod=5)
model_samsung['ma_10'] = ta.SMA(model_samsung.Close, timeperiod=10)
model_samsung['ma_15'] = ta.SMA(model_samsung.Close, timeperiod=15)
model_samsung['ma_20'] = ta.SMA(model_samsung.Close, timeperiod=20)
model_samsung['ma_30'] = ta.SMA(model_samsung.Close, timeperiod=30)
model_samsung['ma_60'] = ta.SMA(model_samsung.Close, timeperiod=60)
model_samsung['ma_120'] = ta.SMA(model_samsung.Close, timeperiod=120)

model_samsung['ema_5'] = ta.EMA(model_samsung.Close, timeperiod=5)
model_samsung['ema_10'] = ta.EMA(model_samsung.Close, timeperiod=10)
model_samsung['ema_15'] = ta.EMA(model_samsung.Close, timeperiod=15)
model_samsung['ema_20'] = ta.EMA(model_samsung.Close, timeperiod=20)
model_samsung['ema_30'] = ta.EMA(model_samsung.Close, timeperiod=30)
model_samsung['ema_60'] = ta.EMA(model_samsung.Close, timeperiod=60)
model_samsung['ema_120'] = ta.EMA(model_samsung.Close, timeperiod=120)

model_samsung['wma_5'] = ta.WMA(model_samsung.Close, timeperiod=5)
model_samsung['wma_10'] = ta.WMA(model_samsung.Close, timeperiod=10)
model_samsung['wma_15'] = ta.WMA(model_samsung.Close, timeperiod=15)
model_samsung['wma_20'] = ta.WMA(model_samsung.Close, timeperiod=20)
model_samsung['wma_30'] = ta.WMA(model_samsung.Close, timeperiod=30)
model_samsung['wma_60'] = ta.WMA(model_samsung.Close, timeperiod=60)
model_samsung['wma_120'] = ta.WMA(model_samsung.Close, timeperiod=120)

model_samsung['ma_v5'] = ta.SMA(model_samsung.Volume, timeperiod=5)
model_samsung['ma_v10'] = ta.SMA(model_samsung.Volume, timeperiod=10)
model_samsung['ma_v20'] = ta.SMA(model_samsung.Volume, timeperiod=20)
model_samsung['ma_v60'] = ta.SMA(model_samsung.Volume, timeperiod=60)
model_samsung['ma_v120'] = ta.SMA(model_samsung.Volume, timeperiod=120)

# 2) 볼린저밴드 (주가의 이동평균선을 중심으로 표준편차 범위를 표시)
ubb, mbb, lbb = ta.BBANDS(model_samsung.Close, 20, 2)
model_samsung['ubb'] = ubb
model_samsung['mbb'] = mbb
model_samsung['lbb'] = lbb

# 3) MACD 이동평균수렴확산 (단기(EMA12)와 장기(EMA26) EMA로 모멘텀을 추정)
macd, macdsignal9, macdhist = ta.MACD(model_samsung.Close, fastperiod=12, slowperiod=26, signalperiod=9)
model_samsung['macd'] = macd
model_samsung['macdsignal9'] = macdsignal9
model_samsung['macdhist'] = macdhist

# 4) RSI 상대강도지수 - 추세의 강도 파악, 과매수, 과매도 국면 판단
model_samsung['rsi'] = ta.RSI(model_samsung.Close, timeperiod=14)

# 5) 스토캐스틱 오늘의 주가가 일정 동안 주가의 변동폭 중에서 어느 정도인 지?
slowk, slowd = ta.STOCH(model_samsung.High, model_samsung.Low, model_samsung.Close, fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
fastk, fastd = ta.STOCHF(model_samsung.High, model_samsung.Low, model_samsung.Close, fastk_period=5, fastd_period=3, fastd_matype=0)
fastk_rsi, fastd_rsi = ta.STOCHRSI(model_samsung.Close, timeperiod=14, fastk_period=5, fastd_period=3, fastd_matype=0)
model_samsung['slowk'] = slowk
model_samsung['slowd'] = slowd
model_samsung['fastk'] = fastk
model_samsung['fastd'] = fastd
model_samsung['fastk_rsi'] = fastk_rsi
model_samsung['fastd_rsi'] = fastd_rsi

# 6) 기타 자주 사용되는 것들
# CCI (Commodity Channel Index), williams'%R, parabolic SAR
# ADX (Average Directional Movement Index)
# plusDI(Plus Directional Indicator), plusDM Plus Directional Movement)
# ATR (Average True Range), OBV (On Balance Volume) 거래량 분석을 통한 주가분석, Variance
model_samsung['cci'] = ta.CCI(model_samsung.High, model_samsung.Low, model_samsung.Close, timeperiod=14)
model_samsung['willR'] = ta.WILLR(model_samsung.High, model_samsung.Low, model_samsung.Close, timeperiod=14)
model_samsung['sar'] = ta.SAR(model_samsung.High, model_samsung.Low, acceleration=0, maximum=0)
model_samsung['adx'] = ta.ADX(model_samsung.High, model_samsung.Low, model_samsung.Close, timeperiod=14)
model_samsung['plus_di'] = ta.PLUS_DI(model_samsung.High, model_samsung.Low, model_samsung.Close, timeperiod=14)
model_samsung['plus_dm'] = ta.PLUS_DM(model_samsung.High, model_samsung.Low, timeperiod=14)
model_samsung['atr'] = ta.ATR(model_samsung.High, model_samsung.Low, model_samsung.Close, timeperiod=14)
model_samsung['obv'] = ta.OBV(model_samsung.Close, model_samsung.Volume)
model_samsung['var'] = ta.VAR(model_samsung.Close, timeperiod=5, nbdev=1)
```

```python
# samsung 모델
lst_label = ['a']
for i in range(len(model_samsung)-1):
    if model_samsung.iloc[i+1]['Close'] == model_samsung.iloc[i]['Close']:
        lst_label.append(2) # 전일 주가 = 당일 주가 : 2
    elif model_samsung.iloc[i+1]['Close'] > model_samsung.iloc[i]['Close']:
        lst_label.append(0) # 전일 주가 < 당일 주가 : 0
    else:
        lst_label.append(1) # 전일 주가 > 당일 주가 : 1
model_samsung['Labeling'] = lst_label
model_samsung = model_samsung.drop(model_samsung[model_samsung['Labeling'] == 'a'].index) # 첫 행 삭제
model_samsung = model_samsung.drop(model_samsung[model_samsung['Labeling'] == 2].index) # 전일 주가 = 당일 주가인 행 삭제
model_samsung['Labeling'] = model_samsung['Labeling'].astype("category")
model_samsung = model_samsung.dropna() # 결측치가 있는 행 제거
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test, pred=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    print('혼동행렬')
    print(confusion)
    # ROC-AUC print 추가
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
```
