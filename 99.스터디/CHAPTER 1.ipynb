# CHAPTER 1 금융과 투자 영역의 머신러닝
---

* 개요 부분으로, 가볍게 읽고 넘어가면 된다.
* AI 기술을 이용한 투자 관리
    * 투자 전략이 공개되면 수익률이 떨어질 위험이 높기 때문에 <span style="font-weight:bold;">학술 논문</span> 외에는 공개되는 일이 드물다.
    * 학술 논문에 게재된 전략이더라도 데이터에 따라 결과가 다르거나 실제 트레이딩에 적용하기 어렵다.
    * 현직 실무자 의견 : AI가 현직 투자자들을 대체할 수는 없지만, 의사결정을 내리는 데에 도움을 줄 수 있음
> AI는 데이터 분석 및 insight를 얻기 위해 사용하고, 실제 의사결정을 내리는 것은 사용자가 해야 함<br/>
> 그러나, 일부 글로벌 기업은 AI 모델에 의지하여 높은 수익률을 올리고 있는 곳도 존재함<br/>
> &emsp; <span style="font-size:0.8em;">※ 책의 33페이지에 나오는 크래프트테크놀로지스는 100% 딥러닝 기반 운용</span><br/>
> 결론적으로, 트레이딩에 있어 AI 기술은 필수불가결하다.


## 1.1 금융, AI, 투자의 삼자관계
----

* 책 29 page, 그림 1-3. 트레이딩 알고리즘을 위해 만들어진 AI 기술
    1. 거시경제 데이터, 재무제표 데이터, 시장 데이터를 입력 변수로 하여 분류 문제로 투자 의사결정 내리기
    2. 가치 시그널, 기술적 시그널, 시장 미시구조 시그널 등을 생성
    3. 비지도학습을 사용하여 자산 간의 관계 파악
    4. 자연어 처리 기술을 사용해 SNS 감정 분석
    5. 시장 트렌드나 레짐(regime) 결정 (은닉 마르코프 모형 활용)  
    <span style="font-size:0.8em;"> 
        &emsp;※ 레짐 : 정권, 체제, 경기 사이클, 시장의 지배적인 분위기<br/>
        &emsp;※ 은닉 마르코프 모델 : 관측치 뒤에 은닉되어 있는 상태(state)를 추정
    </span>
    6. 단기 자산 가격 방향 예측 (라쏘, KNN, 로지스틱 회귀 등)
    7. 거시경제 데이터에 기반해 단기 자산 가격 방향 예측 (그래디언트 부스팅)
    8. 수익률이 좋은 트레이딩 전략 찾기 (강화학습, 딥러닝)
    9. 기존에 알려진 팩터(factor) 분석  
    <span style="font-size:0.8em;"> 
    &emsp;※ 요인(factor)분석 : 변수들 간의 상관관계를 고려하여 서로 유사한 변수들 끼리 묶어주는 방법
    </span>
    10. 노이즈가 섞인 데이터에서 자산 가격의 방향이나 시그널 찾기 (SVM 등 지도학습 분류 알고리즘)

## 1.2 실제 투자 영역의 머신러닝 응용 사례
----

* AI 기술 투자에 활용하기
    1. 거시적인 시장 전체의 움직임(레짐) 예측 → 자산 배분 / 주식 선정하기
    2. 1번과 결합하거나 1번과 관계없이 자산 자체의 움직임 예측
> 글로벌 투자기관에서 실제 market cycle, regime을 예측할 때는<br/> 
> &emsp; * 거시 지표 사용법<br/>
> &emsp; * 자연어 처리 기술을 활용한 패턴 인식 혹은 순환 검출<br/>
> &emsp; * 팩터 기반 접근법<br/>
> 등을 활용한 마켓 시그널로 퀀트나 포트폴리오 매니저에게 정량적 insight를 제공한다.

## 1.3 투자 영역에서 활용하는 알고리즘
---

* 필요지식 : 선형대수, 미적분, 통계학 지식, 컴퓨터 공학, 머신러닝 / 딥러닝 전반
> 알고리즘을 수학적으로 이해하고 관련 프레임워크 사용법을 잘 알아야 한다
* 투자 관련 머신러닝의 강점
    * 샘플 내 분산 조정이 아닌 샘플 외 데이터 예측력
    * 컴퓨터의 계산능력을 활용해 비현실적 가정 회피
    * 비선형 / 계층적 / 고차원의 비연속 상호작용 효과 등 복잡한 내용을 학습 가능
    * 다중공선성과 다른 대체 효과에서 변수 / 상황 탐색 구분 가능  
    <span style="font-size:0.8em;">&emsp;※ 다중공선성 : 회귀 분석 모형의 일부 설명 변수가 다른 설명 변수와 상관 정도가 높아 데이터 분석에 부정적인 영향을 끼치는 경우<br/></span>
    * 페어 트레이딩 등의 전통적인 통계 방법에서 추가적인 데이터를 분석하여 정밀한 전략 구사 가능

## 1.4 투자 영역에서 활용하는 데이터
------

* 책 37페이지, 표 1-2 금융 데이터의 분류
    * 재무제표 데이터 (자산, 부채, 판매량, 비용/이익, 거시 변수, ...)
        * 일 단위의 주가 데이터나 분기별 재무제표 데이터
            * 가장 간단하며, 베이스라인이 되기 때문에 중요
            * <span style="font-weight:bold;">데이터가 적고 노이즈가 많아 과적합 문제가 종종 발생함</span>
    * 마켓 데이터 (가격/변동성, 거래량, 배당, 이자율, 상장/폐지, ...)
        * 거시경제 지표, 장단기 금리 차이, 경기선행지수, GDP, 데이터 밴더의 예측 데이터, 설문 기반의 데이터 등도 활용가능
    * 분석 데이터 (애널리스트 추천, 신용 등급, 이익 예측, 감성 분석, ...)
    * 대체 데이터 (위성/cctv 이미지, 구글 검색어, sns, 메타데이터, ...)
        * 자연어 처리 기술을 활용한 뉴스 / SNS 데이터를 가장 많이 활용

> 대체 데이터의 유용성 검증을 위해서는 다음 6가지를 고려해야 한다  
> &emsp; * 신뢰성 : 데이터 정확도  
> &emsp; * 세분화  
> &emsp; * 효용성 : 최신 트렌드 반영 여부, 수집/가공 후에도 유지되는지  
> &emsp; * 사용 범위  
> &emsp; * 행동가능성 : 데이터 기반 구체적인 의사결정 가능여부  
> &emsp; * 희소성
